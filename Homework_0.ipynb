```python
import cv2
import os
import numpy as np
import matplotlib.pyplot as plt

# Read the image as grayscale
apollo_image = cv2.imread(os.path.relpath('Data/Apollo_11_Launch.jpg'), cv2.IMREAD_GRAYSCALE)

# Display the grayscale image
plt.imshow(apollo_image, cmap="gray")
plt.title("Apollo 11 Launch");
```

```python
# Crop the original image
apollo_image_top = apollo_image[100:200, 600:700];

# Display the cropped, zoomed-in image
plt.imshow(apollo_image_top, cmap="gray");
plt.title("Apollo 11 Launch, Cropped");
```

```python
# Matrix of ones based on the size of the cropped image
matrix = np.ones(apollo_image_top.shape, dtype="uint8") * 125

# Generate the brighter image by adding to each value
apollo_image_top_brighter = cv2.add(apollo_image_top, matrix)

# Generate the darker image by subtracting from each value
apollo_image_top_darker = cv2.subtract(apollo_image_top, matrix)

# Show each of the images
plt.figure(figsize=[18, 5])
plt.subplot(131); plt.imshow(apollo_image_top_darker, cmap="gray");
plt.title("Very Dark");
plt.subplot(132); plt.imshow(apollo_image_top, cmap="gray");
plt.title("Original");
plt.subplot(133); plt.imshow(apollo_image_top_brighter, cmap="gray");
plt.title("Very Bright");
```

```python
# Rotate the cropped, darkened image clockwise by 90 degrees using cv2.rotate()
# (This function doesn't seem to be mentioned in the Notebooks so far, but it seems like what's intended to be used here)
apollo_image_top_darker_rotated = cv2.rotate(apollo_image_top_darker, cv2.ROTATE_90_CLOCKWISE)

# Display and save the final modified image
plt.imshow(apollo_image_top_darker_rotated, cmap="gray");
plt.title("Final Modified Apollo 11 Launch Image");
cv2.imwrite("Data/Apollo_11_Launch_Modified_zmw24.png", apollo_image_top_darker_rotated);
```
